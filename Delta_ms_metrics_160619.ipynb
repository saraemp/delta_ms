```python
Oli_NoFlex=pd.read_csv('./Delta_070518_notect_wall_SL_Oli_NoFlex.csv', sep=" ", header=None)
Oli_Te50=pd.read_csv('./Delta_070518_notect_wall_SL_Oli_Te50_flex.csv', sep=" ", header=None)
Pali_NoFlex=pd.read_csv('./Delta_070518_notect_wall_SL_Pal_NoFlex.csv', sep=" ", header=None)
Pali_Te50=pd.read_csv('./Delta_070518_notect_wall_SL_Pal_Te50_flex.csv', sep=" ", header=None)
```
